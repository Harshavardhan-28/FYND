```python
import pandas as pd

def create_balanced_subset(original_df: pd.DataFrame, n_per_class: int = 40) -> pd.DataFrame:

    clean_df = original_df.drop_duplicates(subset=['text']).copy()


    counts = clean_df['stars'].value_counts()
    if any(counts < n_per_class):
        print(f"WARNING: Not enough data. Counts:\n{counts}")
        return None



    stratified_yelp = clean_df.groupby('stars', group_keys=False).sample(n=n_per_class, random_state=42)


    stratified_yelp = stratified_yelp.sample(frac=1, random_state=42).reset_index(drop=True)

    return stratified_yelp[['review_id', 'text', 'stars']]

stratified_yelp = create_balanced_subset(yelp)


print(f"Subset created. Total Rows: {len(stratified_yelp)}")
print(stratified_yelp['stars'].value_counts().sort_index())
display(stratified_yelp.head(10))
```

```python
from pydantic import BaseModel, Field

class ReviewAnalysis(BaseModel):
    """
    Schema for classifying Yelp reviews. 
    Enforces a strict 1-5 integer rating and a text explanation.
    """
    predicted_stars: int = Field(
        ..., 
        ge=1, 
        le=5, 
        description="The predicted star rating (1, 2, 3, 4, or 5) derived from the sentiment of the review."
    )
    explanation: str = Field(
        ..., 
        description="A brief, 1-2 sentence reasoning justifying the rating based on specific phrases in the text."
    )


try:
    valid_example = ReviewAnalysis(predicted_stars=4, explanation="Food was good.")
    print("Schema Definition is Valid.")
    print(valid_example.model_dump_json(indent=2))
except Exception as e:
    print(f"Schema Error: {e}")
```

```python
import asyncio
import os
import pandas as pd
import numpy as np
from google import genai
from google.genai import types
from tqdm.asyncio import tqdm_asyncio
from sklearn.metrics import accuracy_score, mean_absolute_error, mean_squared_error, classification_report

# --- Configuration ---
# Ensure API key is set
# if "GOOGLE_API_KEY" not in os.environ:
#     from google.colab import userdata
#     os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY')

class GeminiScorer:
    def __init__(self, model_name: str = "gemini-3-flash-preview"):
        # Fixed: explicitly getting key from env
        self.client = genai.Client(api_key=os.getenv("GOOGLE_API_KEY")) 
        self.model_name = model_name
        # Semaphore: 10 concurrent requests is usually safe for Flash tier
        self.semaphore = asyncio.Semaphore(10)

    def _get_zero_shot_prompt(self, review_text: str) -> str:
        """Baseline prompt: Direct instruction, no examples."""
        return f"""
        Role: You are an expert Review Sentiment Analyzer.
        Task: Analyze the provided Yelp review text and determine the star rating (1-5).
        
        Review Text: "{review_text}"
        """

    async def predict_zero_shot(self, review_text: str) -> dict:
        async with self.semaphore:
            try:
                # Using the google-genai SDK v1.0+ structure
                response = await self.client.aio.models.generate_content(
                    model=self.model_name,
                    contents=self._get_zero_shot_prompt(review_text),
                    config=types.GenerateContentConfig(
                        response_mime_type="application/json",
                        response_schema=ReviewAnalysis, # Uses the Pydantic schema from previous cell
                        temperature=0.0 # Deterministic
                    )
                )
                

                result: ReviewAnalysis = response.parsed
                return {
                    "pred_stars": result.predicted_stars,
                    "explanation": result.explanation,
                    "error": None
                }
            except Exception as e:
                return {
                    "pred_stars": None,
                    "explanation": None,
                    "error": str(e)
                }



async def run_zero_shot_evaluation(input_df: pd.DataFrame):
    scorer = GeminiScorer(model_name="gemini-3-flash-preview")
    

    result_df = input_df.copy()
    tasks = []
    
    print(f"üöÄ Starting Zero-Shot Evaluation on {len(result_df)} rows...")
    
    for _, row in result_df.iterrows():
        tasks.append(scorer.predict_zero_shot(row['text']))
    

    results = await tqdm_asyncio.gather(*tasks)
    

    result_df['zs_pred'] = [r['pred_stars'] for r in results]
    result_df['zs_explanation'] = [r['explanation'] for r in results]
    result_df['zs_error'] = [r['error'] for r in results]
    
    return result_df


if 'stratified_yelp' in locals():
    df_zero_shot = await run_zero_shot_evaluation(stratified_yelp)

    valid_results = df_zero_shot.dropna(subset=['zs_pred']).copy()
    y_true = valid_results['stars'].astype(int)
    y_pred = valid_results['zs_pred'].astype(int)


    acc = accuracy_score(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    
    print(f"\nüìä --- Zero-Shot Results ---")
    print(f"‚úÖ Accuracy: {acc:.2%} (Exact Matches)")
    print(f"üìâ MAE:      {mae:.4f} (Avg Error in Stars)")
    print(f"üìê MSE:      {mse:.4f} (Penalizes Big Errors)")
    print(f"‚ùå Errors:   {df_zero_shot['zs_error'].dropna().count()}")


    print("\nüîç Classification Report:")
    print(classification_report(y_true, y_pred, digits=2))


    print("\nüìù Sample Mismatches (Actual != Predicted):")
    mismatches = valid_results[valid_results['stars'] != valid_results['zs_pred']].head(3)
    if not mismatches.empty:
        display(mismatches[['text', 'stars', 'zs_pred', 'zs_explanation']])
else:
    print("‚ùå Error: 'stratified_yelp' dataframe not found. Please run the sampling step.")
```

```python
import pandas as pd
import random

def get_dynamic_few_shot_examples(full_df: pd.DataFrame, test_df: pd.DataFrame) -> list:
    """
    Selects 3 representative examples (1, 3, 5 stars) from the full dataset
    EXCLUDING any rows present in the test set to prevent Data Leakage.
    """
    # 1. Anti-Leakage: Exclude test rows
    # We use review_id as the unique key
    test_ids = set(test_df['review_id'].values)
    train_pool = full_df[~full_df['review_id'].isin(test_ids)].copy()
    
    # 2. Filter for "Prompt Friendly" reviews
    # We want short reviews (< 300 chars) so we don't waste context window
    train_pool = train_pool[train_pool['text'].str.len().between(50, 300)]
    
    examples = []
    
    # 3. Select Anchors (1, 3, 5)
    # We add generic reasoning templates because we can't manually read them all
    reasoning_templates = {
        1: "Critical feedback focusing on negative experience.",
        3: "Mixed feedback indicating an average experience.",
        5: "Highly positive feedback with no significant complaints."
    }
    
    for star in [1, 3, 5]:
        # Sample 1 row
        if len(train_pool[train_pool['stars'] == star]) > 0:
            row = train_pool[train_pool['stars'] == star].sample(n=1, random_state=42).iloc[0]
            examples.append({
                "text": row['text'],
                "stars": row['stars'],
                "reasoning": reasoning_templates[star]
            })
        else:
            raise ValueError(f"Could not find a valid {star}-star example in training pool!")
            
    return examples

# --- Generate the Examples ---
# assuming 'yelp' is your original full dataframe
fs_examples = get_dynamic_few_shot_examples(yelp, stratified_yelp)

print("Selected Few-Shot Examples:")
for i, ex in enumerate(fs_examples):
    print(f"[{i+1}] Stars: {ex['stars']} | Text: {ex['text'][:100]}...")
```

```python
import asyncio
from tqdm.asyncio import tqdm_asyncio
from sklearn.metrics import accuracy_score, mean_absolute_error

class FewShotScorer(GeminiScorer): # Inherits from our previous class
    def __init__(self, examples: list, model_name="gemini-3-flash-preview"):
        super().__init__(model_name)
        self.examples = examples
        
    def _construct_prompt(self, target_text: str) -> str:
        """Builds the prompt with the dynamic examples injected."""
        
        # Build the "History" string
        example_str = ""
        for ex in self.examples:
            example_str += f"""
            Review: "{ex['text']}"
            Rating: {ex['stars']}
            Reasoning: {ex['reasoning']}
            ---
            """
            
        return f"""
        Role: You are an expert Review Sentiment Analyzer.
        
        Instruction: Learn from the following examples of how to rate reviews.
        
        {example_str}
        
        Task: Now analyze the following review and provide the rating and reasoning.
        
        Review: "{target_text}"
        """

    async def predict(self, text: str) -> dict:
        async with self.semaphore:
            try:
                response = await self.client.aio.models.generate_content(
                    model=self.model_name,
                    contents=self._construct_prompt(text),
                    config=types.GenerateContentConfig(
                        response_mime_type="application/json",
                        response_schema=ReviewAnalysis,
                        temperature=0.0
                    )
                )
                result = response.parsed
                return {"fs_pred": result.predicted_stars, "fs_explanation": result.explanation, "fs_error": None}
            except Exception as e:
                return {"fs_pred": None, "fs_explanation": None, "fs_error": str(e)}

# --- Execution ---

async def run_few_shot_evaluation(input_df, examples):
    # 1. Setup
    scorer = FewShotScorer(examples)
    result_df = input_df.copy() # Immutable pattern
    
    print(f"üöÄ Starting Few-Shot (k=3) Evaluation on {len(result_df)} rows...")
    
    # 2. Batch Processing
    tasks = [scorer.predict(row['text']) for _, row in result_df.iterrows()]
    results = await tqdm_asyncio.gather(*tasks)
    
    # 3. Store Results
    result_df['fs_pred'] = [r['fs_pred'] for r in results]
    result_df['fs_explanation'] = [r['fs_explanation'] for r in results]
    result_df['fs_error'] = [r['fs_error'] for r in results]
    
    return result_df

# Run the pipeline
df_few_shot = await run_few_shot_evaluation(stratified_yelp, fs_examples)

# --- Comparison Metrics ---

# Filter valid rows
valid_fs = df_few_shot.dropna(subset=['fs_pred'])
y_true = valid_fs['stars'].astype(int)
y_pred_fs = valid_fs['fs_pred'].astype(int)

# Calculate Stats
acc_fs = accuracy_score(y_true, y_pred_fs)
mae_fs = mean_absolute_error(y_true, y_pred_fs)

print("\nüìä --- Few-Shot Results ---")
print(f"‚úÖ Accuracy: {acc_fs:.2%}")
print(f"üìâ MAE:      {mae_fs:.4f}")

# --- COMPARISON: Zero-Shot vs Few-Shot ---
# We assume 'acc' and 'mae' from the previous zero-shot run are available variables
if 'acc' in locals():
    print("\n‚öñÔ∏è --- STRATEGY COMPARISON ---")
    print(f"Accuracy Delta: {acc_fs - acc:+.2%} ({'Improved' if acc_fs > acc else 'Regressed'})")
    print(f"MAE Delta:      {mae_fs - mae:+.4f} ({'Improved' if mae_fs < mae else 'Regressed'})")
```

```python
# --- IMPROVED STRATEGY 2: Curated Few-Shot ---

import re
# We use the previous class and runner, just changing the Data Input (Examples)

def get_curated_few_shot_examples(full_df: pd.DataFrame, test_df: pd.DataFrame) -> list:
    """
    Selects 'Gold Standard' examples using heuristics rather than pure randomness.
    """
    # 1. Anti-Leakage
    test_ids = set(test_df['review_id'].values)
    train_pool = full_df[~full_df['review_id'].isin(test_ids)].copy()
    
    # 2. Strict Cleaning Filters
    # Remove reviews with digits (avoids "3/5", "10 minutes", "5 stars") to force text analysis
    train_pool = train_pool[~train_pool['text'].str.contains(r'\d', regex=True)]
    # Keep length reasonable (50-200 chars)
    train_pool = train_pool[train_pool['text'].str.len().between(50, 200)]
    
    examples = []
    
    # 3. Define Archetype Heuristics
    heuristics = {
        1: {
            "keywords": ["worst", "horrible", "terrible", "rude", "disgusting"],
            "reasoning": "Strong negative sentiment ('worst', 'horrible')."
        },
        3: {
            "keywords": ["okay", "average", "decent", "mediocre"], 
            "reasoning": "Mixed sentiment. Highlights good and bad points."
        },
        5: {
            "keywords": ["amazing", "best", "perfect", "love", "excellent"],
            "reasoning": "Unconditional positive praise."
        }
    }
    
    for star, rules in heuristics.items():
        # Filter pool for rows with that star AND containing at least one keyword
        keyword_pattern = '|'.join(rules['keywords'])
        candidates = train_pool[
            (train_pool['stars'] == star) & 
            (train_pool['text'].str.lower().str.contains(keyword_pattern))
        ]
        
        if len(candidates) > 0:
            # Pick the shortest one among candidates to save tokens and be punchy
            best_row = candidates.loc[candidates['text'].str.len().idxmin()]
            
            examples.append({
                "text": best_row['text'],
                "stars": star,
                "reasoning": rules['reasoning']
            })
        else:
            # Fallback (should rarely happen in Yelp dataset)
            print(f"‚ö†Ô∏è Warning: No perfect archetype found for {star} stars. Using random fallback.")
            examples.append({
                "text": train_pool[train_pool['stars'] == star].sample(1).iloc[0]['text'],
                "stars": star,
                "reasoning": "Generic example."
            })
            
    return examples

# --- EXECUTION ---

# 1. Get New Examples
curated_examples = get_curated_few_shot_examples(yelp, stratified_yelp)

print("‚ú® Curated 'Golden' Examples:")
for ex in curated_examples:
    print(f"[{ex['stars']} Star] {ex['text']}")

# 2. Re-Run Evaluation (Version 2)
print("\nüîÑ Re-running Few-Shot with Curated Examples...")

# We reuse the function from the previous cell, passing the NEW examples
df_few_shot_v2 = await run_few_shot_evaluation(stratified_yelp, curated_examples)

# 3. Calculate New Metrics
valid_fs_v2 = df_few_shot_v2.dropna(subset=['fs_pred'])
acc_v2 = accuracy_score(valid_fs_v2['stars'], valid_fs_v2['fs_pred'])
mae_v2 = mean_absolute_error(valid_fs_v2['stars'], valid_fs_v2['fs_pred'])

print(f"\nüìä --- Improved Few-Shot Results (V2) ---")
print(f"‚úÖ Accuracy: {acc_v2:.2%}")
print(f"üìâ MAE:      {mae_v2:.4f}")

# 4. Compare with Zero Shot (Baseline)
# Assuming 'acc' variable from Zero Shot step still exists
if 'acc' in locals():
    delta = acc_v2 - acc
    print(f"Accuracy Delta vs Zero-Shot: {delta:+.2%} ({'Improved' if delta > 0 else 'Regressed'})")
```

```python
# --- STRATEGY 3: RUBRIC + CHAIN OF THOUGHT (CoT) ---

class RubricCoTScorer(GeminiScorer):
    def __init__(self, model_name="gemini-3-flash-preview"):
        super().__init__(model_name)
        
    def _get_rubric_prompt(self, text: str) -> str:
        """
        Injects a strict grading rubric and forces 'Step-by-Step' reasoning.
        """
        rubric = """
        GRADE USING THIS STRICT RUBRIC:
        * 1 Star (Unacceptable): Dealbreakers present (Food poisoning, pests, rude/hostile staff, raw food).
        * 2 Stars (Poor): Major negatives (Cold food, hour+ wait) that overshadow any positives.
        * 3 Stars (Average): "Okay" or "Meh". Edible food but forgettable service. No strong feelings.
        * 4 Stars (Very Good): Great experience but with ONE minor complaint (e.g., "noisy", "hard to park", "expensive").
        * 5 Stars (Exceptional): Flawless execution. "Best ever". No negative caveats mentioned.
        """
        
        return f"""
        Role: You are a strict Quality Assurance Auditor for Yelp.
        
        {rubric}
        
        Task:
        1. Read the review below.
        2. Identify PROS and CONS.
        3. Match the specific combination of pros/cons to the RUBRIC levels above.
        4. Assign the rating.
        
        Review Text: "{text}"
        """

    async def predict(self, text: str) -> dict:
        async with self.semaphore:
            try:
                # We use the same schema, but the prompt changes how the model generates the data
                response = await self.client.aio.models.generate_content(
                    model=self.model_name,
                    contents=self._get_rubric_prompt(text),
                    config=types.GenerateContentConfig(
                        response_mime_type="application/json",
                        response_schema=ReviewAnalysis,
                        temperature=0.0
                    )
                )
                result = response.parsed
                return {"cot_pred": result.predicted_stars, "cot_explanation": result.explanation, "cot_error": None}
            except Exception as e:
                return {"cot_pred": None, "cot_explanation": None, "cot_error": str(e)}

# --- EXECUTION ---

async def run_rubric_evaluation(input_df):
    scorer = RubricCoTScorer()
    result_df = input_df.copy()
    
    print(f"üöÄ Starting Rubric/CoT Evaluation on {len(result_df)} rows...")
    
    tasks = [scorer.predict(row['text']) for _, row in result_df.iterrows()]
    results = await tqdm_asyncio.gather(*tasks)
    
    result_df['cot_pred'] = [r['cot_pred'] for r in results]
    result_df['cot_explanation'] = [r['cot_explanation'] for r in results]
    result_df['cot_error'] = [r['cot_error'] for r in results]
    
    return result_df

# 1. Run Strategy 3
df_rubric = await run_rubric_evaluation(stratified_yelp)

# 2. Calculate Metrics
valid_cot = df_rubric.dropna(subset=['cot_pred'])
acc_cot = accuracy_score(valid_cot['stars'], valid_cot['cot_pred'])
mae_cot = mean_absolute_error(valid_cot['stars'], valid_cot['cot_pred'])

print(f"\nüìä --- Strategy 3: Rubric + CoT Results ---")
print(f"‚úÖ Accuracy: {acc_cot:.2%}")
print(f"üìâ MAE:      {mae_cot:.4f}")

# 3. Final Comparison (Assuming 'acc' from ZeroShot and 'acc_v2' from FewShot exist)
if 'acc' in locals() and 'acc_v2' in locals():
    print("\nüèÜ FINAL LEADERBOARD üèÜ")
    print(f"1. Zero-Shot:  {acc:.2%}")
    print(f"2. Few-Shot:   {acc_v2:.2%}")
    print(f"3. Rubric/CoT: {acc_cot:.2%}")
    
    best_strategy = max([("Zero-Shot", acc), ("Few-Shot", acc_v2), ("Rubric", acc_cot)], key=lambda x: x[1])
    print(f"\nü•á Winner: {best_strategy[0]}")
```

```python
# --- STRATEGY 3.1: Structured Chain-of-Thought ---

# 1. New Schema with "Thinking Fields"
class ReviewWithReasoning(BaseModel):
    """
    Forces the model to list pros/cons BEFORE assigning a rating.
    This acts as a 'Scratchpad' for the model.
    """
    key_positives: str = Field(..., description="List of positive points mentioned.")
    key_negatives: str = Field(..., description="List of negative points mentioned.")
    # We relax the constraint slightly to match human irrationality
    predicted_stars: int = Field(..., ge=1, le=5) 
    explanation: str = Field(..., description="Final justification.")

class StructuredCoTScorer(GeminiScorer):
    def __init__(self, model_name="gemini-3-flash-preview"):
        super().__init__(model_name)
        
    def _get_reasoning_prompt(self, text: str) -> str:
        return f"""
        Role: You are an empathetic Review Analyzer. You understand that customers sometimes rate high even with flaws, or low for minor reasons.
        
        Task:
        1. Extract the positives and negatives into the JSON fields.
        2. Weigh them as the *user* likely did.
        3. Assign the star rating that best matches the user's tone.
        
        Review Text: "{text}"
        """

    async def predict(self, text: str) -> dict:
        async with self.semaphore:
            try:
                response = await self.client.aio.models.generate_content(
                    model=self.model_name,
                    contents=self._get_reasoning_prompt(text),
                    config=types.GenerateContentConfig(
                        response_mime_type="application/json",
                        response_schema=ReviewWithReasoning, # NEW SCHEMA
                        temperature=0.0
                    )
                )
                result = response.parsed
                return {
                    "cot_pred": result.predicted_stars, 
                    "cot_explanation": result.explanation, 
                    "cot_error": None
                }
            except Exception as e:
                return {"cot_pred": None, "cot_explanation": None, "cot_error": str(e)}

# --- EXECUTION ---

async def run_structured_cot(input_df):
    scorer = StructuredCoTScorer()
    result_df = input_df.copy()
    
    print(f"üöÄ Starting Structured CoT (Internal Scratchpad) on {len(result_df)} rows...")
    
    tasks = [scorer.predict(row['text']) for _, row in result_df.iterrows()]
    results = await tqdm_asyncio.gather(*tasks)
    
    result_df['cot_pred'] = [r['cot_pred'] for r in results]
    result_df['cot_explanation'] = [r['cot_explanation'] for r in results]
    result_df['cot_error'] = [r['cot_error'] for r in results]
    
    return result_df

# 1. Run Strategy 3.1
df_cot_v2 = await run_structured_cot(stratified_yelp)

# 2. Metrics
valid_cot_v2 = df_cot_v2.dropna(subset=['cot_pred'])
acc_cot_v2 = accuracy_score(valid_cot_v2['stars'], valid_cot_v2['cot_pred'])
mae_cot_v2 = mean_absolute_error(valid_cot_v2['stars'], valid_cot_v2['cot_pred'])

print(f"\nüìä --- Strategy 3.1: Structured CoT Results ---")
print(f"‚úÖ Accuracy: {acc_cot_v2:.2%}")
print(f"üìâ MAE:      {mae_cot_v2:.4f}")

# 3. Final Leaderboard
if 'acc' in locals() and 'acc_v2' in locals():
    print("\nüèÜ UPDATED LEADERBOARD üèÜ")
    print(f"1. Zero-Shot:      {acc:.2%}")
    print(f"2. Few-Shot (V2):  {acc_v2:.2%}")
    print(f"3. Strict Rubric:  {acc_cot:.2%} (Failed)")
    print(f"4. Structured CoT: {acc_cot_v2:.2%}")
```

```python
# --- STRATEGY 4 (REVISED): Self-Refinement / Reflective CoT ---

class ReviewReflection(BaseModel):
    """
    Forces the model to double-check its own logic.
    """
    initial_impression: int = Field(..., description="The rating you would give based on a first quick read.")
    critique: str = Field(..., description="Look for evidence that contradicts your initial impression. (e.g., 'Initially thought 5, but text mentions dirty floors').")
    final_reasoning: str = Field(..., description="Synthesis of the impression and critique.")
    final_stars: int = Field(..., ge=1, le=5, description="The final, corrected star rating.")

class ReflectiveScorer(GeminiScorer):
    def __init__(self, model_name="gemini-3-flash-preview"):
        super().__init__(model_name)
        
    def _get_reflective_prompt(self, text: str) -> str:
        # CLEAN PROMPT: No complex JSON examples to confuse the parser.
        # Pure logic instruction.
        return f"""
        Role: You are a meticulous Senior Editor.
        
        Task: 
        1. Read the review.
        2. Form an initial rating.
        3. CRITIQUE that rating. specificially look for "dealbreakers" (1-star signals) or "minor complaints" (drops 5 to 4).
        4. Adjust and finalize the score.
        
        Review: "{text}"
        """

    async def predict(self, text: str) -> dict:
        async with self.semaphore:
            try:
                response = await self.client.aio.models.generate_content(
                    model=self.model_name,
                    contents=self._get_reflective_prompt(text),
                    config=types.GenerateContentConfig(
                        response_mime_type="application/json",
                        response_schema=ReviewReflection, # New Schema
                        temperature=0.0
                    )
                )
                result = response.parsed
                # We return the FINAL stars, not the initial one
                return {
                    "ref_pred": result.final_stars, 
                    "ref_initial": result.initial_impression,
                    "ref_critique": result.critique,
                    "ref_error": None
                }
            except Exception as e:
                return {"ref_pred": None, "ref_initial": None, "ref_critique": None, "ref_error": str(e)}

# --- EXECUTION ---

async def run_reflective_evaluation(input_df):
    scorer = ReflectiveScorer()
    result_df = input_df.copy()
    
    print(f"üöÄ Starting Strategy 4 (Self-Refinement) on {len(result_df)} rows...")
    
    tasks = [scorer.predict(row['text']) for _, row in result_df.iterrows()]
    results = await tqdm_asyncio.gather(*tasks)
    
    result_df['ref_pred'] = [r['ref_pred'] for r in results]
    result_df['ref_initial'] = [r['ref_initial'] for r in results]
    result_df['ref_critique'] = [r['ref_critique'] for r in results]
    result_df['ref_error'] = [r['ref_error'] for r in results]
    
    return result_df

# 1. Run Revised Strategy 4
df_reflective = await run_reflective_evaluation(stratified_yelp)

# 2. Metrics
valid_ref = df_reflective.dropna(subset=['ref_pred'])

# Safety Check: If we still have errors, print the first one
if len(valid_ref) < len(df_reflective):
    print(f"‚ö†Ô∏è Warning: {len(df_reflective) - len(valid_ref)} errors occurred.")
    print("Sample Error:", df_reflective['ref_error'].dropna().iloc[0])

# Stats
y_true = valid_ref['stars'].astype(int)
y_pred = valid_ref['ref_pred'].astype(int)

acc_ref = accuracy_score(y_true, y_pred)
mae_ref = mean_absolute_error(y_true, y_pred)

print(f"\nüìä --- Strategy 4 (Revised): Self-Refinement Results ---")
print(f"‚úÖ Accuracy: {acc_ref:.2%}")
print(f"üìâ MAE:      {mae_ref:.4f}")

# 3. Did "Thinking" help?
# Check how many times the model changed its mind (Initial != Final)
changed_mind = valid_ref[valid_ref['ref_pred'] != valid_ref['ref_initial']]
print(f"üí° The model 'Corrected' itself in {len(changed_mind)} cases.")
if not changed_mind.empty:
    print("Sample Correction:")
    display(changed_mind[['text', 'stars', 'ref_initial', 'ref_pred', 'ref_critique']].head(2))

# --- FINAL COMPARISON TABLE ---
# Consolidating all results into one clean DataFrame
results_summary = {
    "Strategy": ["Zero-Shot", "Few-Shot (Curated)", "Structured CoT", "Self-Refinement"],
    "Accuracy": [
        acc if 'acc' in locals() else 0, 
        acc_v2 if 'acc_v2' in locals() else 0, 
        acc_cot_v2 if 'acc_cot_v2' in locals() else 0, 
        acc_ref
    ],
    "MAE": [
        mae if 'mae' in locals() else 0, 
        mae_v2 if 'mae_v2' in locals() else 0, 
        mae_cot_v2 if 'mae_cot_v2' in locals() else 0, 
        mae_ref
    ]
}

final_df = pd.DataFrame(results_summary).sort_values("Accuracy", ascending=False)

print("\nüèÜ FINAL PROJECT RESULTS üèÜ")
display(final_df.style.background_gradient(cmap='Greens', subset=['Accuracy'])
                  .background_gradient(cmap='Reds_r', subset=['MAE']))
```
